```python
import pandas as pd
import plotly.express as px
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

import os
import re
```

```python
raw_data  = pd.read_excel('./RawData\сводные_4_курс\\сводные_4_курс\\2020-2021\\1\\Excel\\Сводная ведомость 119-1.xlsx')

replace_dict = {
    ord('\n'): ' ',
    ord(','): '',
    ord('.'): '',
    ord('/'): ' ',
}



def df_preprocessing(raw_data: pd.DataFrame, replace_dict: dict):
    raw_data = raw_data.iloc[7::]
    raw_data.iloc[0, 2:15] = raw_data.iloc[1, 2:15]
    raw_data = pd.concat([raw_data.iloc[0:1], raw_data.iloc[2::]], axis=0)
    raw_data = raw_data.dropna(axis=0, thresh=2)
    raw_data = raw_data.iloc[0:-1]
    raw_data.columns = raw_data.iloc[0]
    raw_data.columns = [' '.join(s.translate(replace_dict).split()).lower() for s in raw_data.columns]
    raw_data = raw_data[1::]
    raw_data = raw_data.drop('№', axis=1)
    raw_data.reset_index(drop=True, inplace=True)
    raw_data.columns.name = "id"

    return raw_data

raw_data = df_preprocessing(raw_data, replace_dict)

cols_119_1 = raw_data.columns
raw_data
```

```python
directory = r"G:\Work\College\DataLesson\RawData\сводные_4_курс\сводные_4_курс\2020-2021\1\Excel"
dfs = []

# Проход по всем файлам в директории
for filename in os.listdir(directory):
    # Проверка, содержит ли имя файла строку "119" и является ли файл Excel
    if "119" in filename and filename.endswith('.xlsx'):
        # Полный путь к файлу
        file_path = os.path.join(directory, filename)
        # Открытие файла как pandas DataFrame
        df = pd.read_excel(file_path)
        df = df_preprocessing(df, replace_dict)
        df.columns = cols_119_1
        dfs.append(df)

combined_df_1_1  = pd.concat(dfs, ignore_index=True)
combined_df_1_1
```

```python
raw_data_1_2  = pd.read_excel('./RawData\сводные_4_курс\\сводные_4_курс\\2020-2021\\2\\Сводная ведомость 119-1.xlsx')

def df_preprocessing_1_2(raw_data: pd.DataFrame, replace_dict: dict):
    raw_data = raw_data.iloc[7::]
    raw_data.iloc[0, 2:15] = raw_data.iloc[1, 2:15]
    raw_data = pd.concat([raw_data.iloc[0:1], raw_data.iloc[2::]], axis=0)
    raw_data.columns = raw_data.iloc[0]
    raw_data = raw_data[1::]
    raw_data = raw_data.dropna(axis=0, thresh=2)
    raw_data = raw_data.iloc[0:-1]
    raw_data = raw_data.drop('№', axis=1)
    raw_data.reset_index(drop=True, inplace=True)
    raw_data.columns.name = "id"
    raw_data.rename(columns={raw_data.columns[13]: "Бюджет/контракт"}, inplace=True)
    raw_data = raw_data.drop(raw_data.columns[-4::], axis=1)
    raw_data.columns = [' '.join(s.translate(replace_dict).split()).lower() for s in raw_data.columns]
    
    return raw_data

raw_data_1_2 = df_preprocessing_1_2(raw_data_1_2, replace_dict)
cols_119_2 = raw_data_1_2.columns
raw_data_1_2.head()
```

```python
directory = r"G:\Work\College\DataLesson\RawData\сводные_4_курс\сводные_4_курс\2020-2021\2"
dfs = []

# Проход по всем файлам в директории
for filename in os.listdir(directory):
    # Проверка, содержит ли имя файла строку "119" и является ли файл Excel
    if "119" in filename and filename.endswith('.xlsx'):
        # Полный путь к файлу
        file_path = os.path.join(directory, filename)
        # Открытие файла как pandas DataFrame
        df = pd.read_excel(file_path)
        df = df_preprocessing_1_2(df, replace_dict)
        df.columns = [i+"_2_семестр" for i in cols_119_2]
        dfs.append(df)

combined_df_1_2  = pd.concat(dfs, ignore_index=True)
combined_df_1_2
```
